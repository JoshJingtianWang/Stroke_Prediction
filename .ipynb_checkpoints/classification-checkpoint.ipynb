## Preprocessing pipeline

### Custom Preprocessor KNN Imputer for mixed data types

```python
class CustomPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
        # self.cat_cols = None
        # self.cont_cols = None
        # self.smoke_cols = None
    
    def get_cols(self, X):

        self.cat_cols = X.select_dtypes('object').columns.tolist()
        self.cat_cols.remove('smoking_status')
        self.cont_cols = X.select_dtypes('number').columns.tolist()
        self.smoke_cols = ['smoking_status']

        self.onehotencoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

        self.ct1 = ColumnTransformer([
            ('cat', self.onehotencoder, self.cat_cols),
            ('num', MinMaxScaler(), self.cont_cols),
            ('smoke', OrdinalEncoder(), self.smoke_cols)],
            remainder='passthrough'
        )

        self.ct2 = ColumnTransformer([
            ('smoke', self.onehotencoder, [-1])],
            remainder='passthrough'
        )

        self.preprocessor = Pipeline([
            ('ct1', self.ct1),
            ('imputer', KNNImputer(n_neighbors=1)),
            ('ct2', self.ct2)
        ])
    
    def fit(self, X, y=None):
        
        self.get_cols(X)
        
        return self.preprocessor.fit(X)

    def transform(self, X, y=None):
        
        #self.fit(X)
        
        return self.preprocessor.transform(X)

    def fit_transform(self, X, y=None, **fit_params):
        
        self.get_cols(X)
        
        return self.preprocessor.fit_transform(X)
    
    def get_feature_names_out(self):
        return self.preprocessor.get_feature_names_out()
    
    def get_smoke_columns(self, X):
        
        self.ordencoder = OrdinalEncoder()
        self.ordencoder.fit(X.smoking_status.to_numpy().reshape(-1,1))
        #self.onehotencoder.fit(self.preprocessor2.fit_transform(X)[-1].reshape(-1,1))
        self.keys = self.ordencoder.categories_
        self.vals = self.ordencoder.transform(np.array(self.ordencoder.categories_).reshape(-1,1))
        
        return dict(zip(self.vals[:,0],self.keys[0]))

preprocessor = CustomPreprocessor()
```

# Defining the evaluation metric: Brier Skill Score
https://en.wikipedia.org/wiki/Brier_score#Brier_Skill_Score_(BSS)

```python
y_train.value_counts()
```

```python
probabilities = [199/(3888+199) for _ in range(len(y_train))]
brier_ref = brier_score_loss(y_train, probabilities) #baseline prediction probability
def brier_skill_score(y_true, y_predicted):
    # calculate the brier score
    bs = brier_score_loss(y_true, y_predicted)
    # calculate skill score
    return 1.0 - (bs / brier_ref)

#Brier skill score. Higher is better
BSS_scorer = make_scorer(brier_skill_score, greater_is_better=True, needs_proba=True)
```

# Logistic Regression

## Base logistic regression

```python
#Logistic Regression Base
preprocessor = CustomPreprocessor()

lr = LogisticRegression(max_iter=200)

cv_pipe = Pipeline([
('preprocessor', preprocessor),
('lr', lr)])
```

```python
skf = StratifiedKFold(n_splits=5)

scores = cross_validate(cv_pipe, X_train, y_train, cv=skf, scoring=(BSS_scorer), return_estimator=True)

#mean BSS of lr base
np.mean(scores['test_score'])
```

## Tuned Logistic Regression

```python
#Logistic Regression GridSearch
preprocessor = CustomPreprocessor()

ratios=[]
for n in np.arange(1,6):
    # setup for the ratio argument of RandomOverSampler initialization
    n_pos = np.sum(y_train == 1)
    n_neg = np.sum(y_train == 0)
    ratios.append({1 : n_pos * n, 0 : n_neg})

classweights=[]
for n in np.arange(1,11,3):
    classweights.append({1 : n, 0 : 1})

penalty=['l1', 'l2'] 
    
lr_pipe = Pipeline([
        ('preprocessor', preprocessor),
        #('poly', polynomial),
        #('oversampling', over_sampling.SMOTE()),
        #('undersampling', under_sampling.NeighbourhoodCleaningRule()),
        ('lr', LogisticRegression(solver='liblinear'))
        ])

params = [{
         #'oversampling__sampling_strategy':ratios,
         #'oversampling__random_state':[42],
#         'undersampling__sampling_strategy':['majority'],
#         'undersampling__n_neighbors':np.arange(1,5),
        'lr__C': np.logspace(0.001, 4, 4),
        'lr__class_weight':classweights,
        'lr__penalty':penalty
        }]

skf = StratifiedKFold(n_splits=5)

gs_lr = GridSearchCV(lr_pipe,
                    param_grid=params,
                    scoring=BSS_scorer,
                    refit=True,
                    cv=skf,
                    n_jobs=-1,
                    error_score='raise',
                    verbose=4)

gs_lr.fit(X_train, y_train)
#y_pred_gs=gs_lr.predict(X_test)

print(gs_lr.best_params_)
print(gs_lr.best_score_)
```
